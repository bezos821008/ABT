```python
import ccxt
import pandas as pd
```

```python
exchange = ccxt.binance({
    'enableRateLimit': True,
})
ba_response = exchange.fapiPublicGetPremiumIndex()
df = pd.DataFrame(ba_response)
df['nextFundingTime'] = pd.to_datetime(df['nextFundingTime'],unit='ms')
df['time'] = pd.to_datetime(df['time'],unit='ms')
df_1 = df[['symbol','time','nextFundingTime']]
df_2 = df[['lastFundingRate']]
```

```python
for i in range(df_2.shape[0]):
    df_2.iloc[i][0] = round(float(df_2.iloc[i][0])*100,3)
```

```python
df3 = df_1.merge(df_2, how='left', left_index=True, right_index=True)
```

```python
pd_csv = pd.read_csv(r'C:\Users\Chiang\Virtual_Coin\Currency_Ranking.csv')
pd_csv.rename(columns={'0':'symbol'})
for i in range(pd_csv.shape[0]):
    pd_csv.iloc[i][0] = pd_csv.iloc[i][0] + 'USDT'
```

```python
pd_csv = pd_csv.rename(columns={'0':'symbol'})
```

```python
Currency_List = pd_csv.head(50)
```

```python
df4 = pd.merge(Currency_List,df3,how='inner')
```
