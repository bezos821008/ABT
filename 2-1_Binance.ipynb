```python
import ccxt
import pandas as pd
import os
import csv
```

```python
exchange = ccxt.binance({
    'enableRateLimit': True,
})
ba_response = exchange.fapiPublicGetPremiumIndex()
```

```python
Currency_List = []
```

```python
for i in ba_response:
    Currency_List.append(i['symbol'])
    print(i['symbol'])
    print(float(i['lastFundingRate'])*100)
    #print("")
```

```python
def csv_function(exchange_name,currency_list):
    
    filepath = os.getcwd() + r'\Coin_SQL'
    if os.path.exists(filepath) == True:
        cot = 0
        for i in range(len(currency_list)):
            cot = cot+1
            if cot <9999:
                symbol = currency_list[i]
                filepath = os.getcwd() + r'\Coin_SQL\\' + symbol 

                if os.path.exists(filepath) == False:
                    with open(filepath, 'a', newline='') as csv_write:
                        writer = csv.writer(csv_write, delimiter=',')
                        writer.writerow([exchange_name])

                if os.path.exists(filepath) == True:
                    with open(filepath, 'r', newline='') as csv_read:
                        rows = list(csv.reader(csv_read))
                        if [exchange_name] not in rows:
                            with open(filepath, 'a', newline='') as csv_write:
                                    writer = csv.writer(csv_write, delimiter=',')
                                    writer.writerow([exchange_name])
                                    #print((rows))
```

```python
csv_function('binance',Currency_List)
```

```python
def funding_rate_binance():
    binance = ccxt.bitfinex({'options': {
        'defaultType': 'future',
    }})    

    fund = binance.parse_funding_rate(symbol='BTC/USD')
    return fund['interestRate']
```
