```python
markets = []
for i in ccxt.bybit().load_markets().keys():
    str1 = i.split('/')
    if len(str1)==2:
        markets.append(i)
        
currency_list = exchange.fetchTickers(markets)
```

```python
for dic in currency_list:
    print(currency_list[dic]['info']['symbol'])
    print(float(currency_list[dic]['info']['funding_rate'])*100)
    #print(currency_list[dic])
    print("")
```
