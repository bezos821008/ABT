```python
# coding: utf-8
import requests

host = "https://api.gateio.ws"
prefix = "/api/v4"
headers = {'Accept': 'application/json', 'Content-Type': 'application/json'}

url = '/futures/usdt/contracts'
query_param = ''
r = requests.request('GET', host + prefix + url, headers=headers)
print(r.json())
```

```python
for i in r.json():
    print(i['name'])
    #print(i['funding_rate_indicative'])
    print(float(i['funding_rate'])*100)
    print("")
```
